## Part I:

**Create a team notebook to read in the COVID-19 data (cases, deaths, and population) using pandas and display the dataframe in a notebook. Merge all the three variables (cases, deaths, and population) to create a super COVID-19 datafame. Export it to a csv format.**

Import pandas and use it to read in the three csv files.

```python
import pandas as pd
```

```python
cases = pd.read_csv("../../../data/stage_I/covid_confirmed_usafacts.csv", delimiter=",")
population = pd.read_csv("../../../data/stage_I/covid_county_population_usafacts.csv", delimiter=",")
deaths = pd.read_csv("../../../data/stage_I/covid_deaths_usafacts.csv", delimiter=",")
```

Display all three dataframes.

```python
cases
```

```python
deaths
```

```python
population
```

Merge the three datasets using countyFIPS and State as the common variables to join.

```python
casesDeaths = cases.merge(deaths, left_on = ['countyFIPS', 'State'], right_on = ['countyFIPS', 'State'], suffixes=('_cases', '_deaths'))
finalDataset = casesDeaths.merge(population, left_on = ['countyFIPS', 'State'], right_on = ['countyFIPS', 'State'])
```

Reorder the dataset to make it more readable.

```python
finalDataset = finalDataset[['countyFIPS', 'County Name', 'State', 'StateFIPS_cases', 'population'] + sorted(list(finalDataset.columns))[:-8]]
```

```python
finalDataset = finalDataset.rename(columns={'StateFIPS_cases': 'StateFIPS'})
```

Display the merged dataset.

```python
finalDataset
```

Save the final dataset to a csv file.

```python
finalDataset.to_csv("../../../data/stage_I/superDataset.csv",index=False)
```
