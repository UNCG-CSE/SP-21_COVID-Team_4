Import pandas and read in the super dataset created in the team program.

```python
import pandas as pd

superDataset = pd.read_csv("../../../../data/stage_I/superDataset.csv", delimiter=",")
```

Create a dataframe with information from Oregon state only.

```python
oregonData = superDataset.loc[superDataset['State'] == 'OR']
```

Create a dataframe with information from Oregon only and the cases and deaths for the past week only.

```python
lastWeek = oregonData[oregonData.columns[-14:]]
```

Calculate the total sum of cases in the state for each day of the last week. Then subtract the value of the previous day to get the number of new cases for the state each day. You can see that the cases were stable at the beginning of the week, had a big spike mid week and then trend mostly downwards for the rest of the week.

```python
lastWeekCases = lastWeek[lastWeek.columns[::2]]
totalCases = lastWeekCases.sum(axis=0)
totalCases = totalCases.subtract(totalCases[0])

totalCases.diff()
```

Do the same process for deaths and see that the number of deaths was stable at the beginning of the week. Then there was a spike mid week and the deaths trend mostly downwards from there.

```python
lastWeekDeaths = lastWeek[lastWeek.columns[1::2]]
totalDeaths = lastWeekDeaths.sum(axis=0)
totalDeaths = totalDeaths.subtract(totalDeaths[0])

totalDeaths.diff()
```

Read in the presidential election csv file and display it.

```python
presidentialElec = pd.read_csv("../../../../data/stage_I/president_county_candidate.csv", delimiter=",")

presidentialElec
```

Read in the state name and abbreviation csv to link the super dataset with the presidential election dataset.

```python
stateAbbr = pd.read_csv("../../../../data/stage_I/name-abbr.csv", delimiter=",", header=None)
```

Merge the presidential election dataset with the state names and abbreviations dataset on the value state.

```python
presDataset = presidentialElec.merge(stateAbbr, left_on = 'state', right_on = 0)
presDataset
```

Merge the super dataset with the presidential election dataset on the values of state and county. Display the final dataset.

```python
presidentialSuperDataset = superDataset.merge(presDataset, left_on = ['State', 'County Name'], right_on = [1, 'county'])
presidentialSuperDataset
```
