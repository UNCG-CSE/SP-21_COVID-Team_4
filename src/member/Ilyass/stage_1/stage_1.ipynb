```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Import and show hospital bed data


```python
#show hospital bed data
HospitalBedData = pd.read_csv("../../../../data/stage_I/Definitive_Healthcare _USA_Hospital_Beds.csv")
HospitalBedData.head()
```

# Import and show covid super dataset

```python
#show covid super dataset
SuperData = pd.read_csv("../../../../data/stage_I/superDataset.csv")
SuperData.head()
```


# Pull all data for Texas only
### Im choosing the state of Texas to focus on and analyze the trends in the last week for, below I am pulling all entries of Texas


```python
#extract all data for Texas only
TexasData = SuperData.loc[(SuperData["State"] == "TX") & (SuperData["countyFIPS"] != 0)]
TexasData.head()
```

# Extract last weeks cases numbers in Texas
### For ease am I separating deaths and cases to show the trends, below am I extracting the cases for the last week

```python
#seperate cases from deaths
TexasDataLastWeekCases = TexasData[["2021-02-11_cases","2021-02-10_cases","2021-02-09_cases","2021-02-08_cases","2021-02-07_cases","2021-02-06_cases","2021-02-05_cases"]]
TexasDataLastWeekCases.head()
```

# Extract last weeks death numbers in Texas
### For ease am I separating deaths and cases to show the trends, below am I extracting the deaths for the last week

```python
#seperate deaths from cases
TexasDataLastWeekDeaths = TexasData[["2021-02-11_deaths","2021-02-10_deaths","2021-02-09_deaths","2021-02-08_deaths","2021-02-07_deaths","2021-02-06_deaths","2021-02-05_deaths"]]
TexasDataLastWeekDeaths.head()
```

# Sum up all the deaths for all counties in Texas
### All the deaths from all countries for each day are summed up, this will allow a trend to be seen, when plotted. 

```python
#Extract and sum death data
TexasDeathSum = pd.DataFrame(TexasDataLastWeekDeaths,columns=['2021-02-11_deaths','2021-02-10_deaths','2021-02-09_deaths','2021-02-08_deaths','2021-02-07_deaths','2021-02-06_deaths','2021-02-05_deaths'])
TexasDeathSumFinal = TexasDeathSum.sum(axis=0)
TexasDeathSumFinal.head()
```

### Flip the data, becuase if it were currently ploted the graph would be unnatrual and misleading since the days would be backwards from the norm of what people expect and would be hard to read

```python
#flip data
FlippedTexasDeathsSum = TexasDeathSumFinal.iloc[::-1]
```

### plot the commbined deaths data to show a trend

```python
#plot
plt.title('Death Trend In Texas Last Week')
plt.xlabel('Days')
plt.ylabel('Deaths')
FlippedTexasDeathsSum.plot( figsize=(15,7), linewidth=4)
```

# From the graph above it can be seen the trend with covid deaths in Texas over the last week had been up, its increaesing by a good amount at about 250 incease every day for the past week, and doesnt show signs of going down

# Sum up all the cases for all counties in Texas
### All the cases from all countries for each day are summed up, this will allow a trend to be seen, when plotted. 

```python
#extract and sum last weeks of cases
TexasCaseSum = pd.DataFrame(TexasDataLastWeekCases,columns=["2021-02-11_cases","2021-02-10_cases","2021-02-09_cases","2021-02-08_cases","2021-02-07_cases","2021-02-06_cases","2021-02-05_cases"])
TexasCaseSumFinal = TexasCaseSum.sum(axis=0)
TexasCaseSumFinal.head()
```

### Flip the data, becuase if it were currently ploted the graph would be unnatrual and misleading since the days are would be backwards from the norm of what people expect and would be hard to read

```python
#flip data
FlippedTexasCaseSum = TexasCaseSumFinal.iloc[::-1]
```

### plot the combined deaths data to show a trend

```python
#plot 
plt.title('Case Trend In Texas Last Week')
plt.xlabel('Days')
plt.ylabel('Cases')
FlippedTexasCaseSum.plot( figsize=(15,7), linewidth=4)
```

# From the graph above it can be seen the trend with covid cases in Texas over the last week had been up, its increaesing by a good amount every day for the past week, and doesnt show signs of going down, the graph shows almost a straight line 

### Below the hospital bed data is being shaped into a better more managble form before merging with the covid superset
#### Some useless coulms are being removed, NANs are being removed and normailzed, and data with floating points are being converted into ints to make merging and summing easier

```python
#drop columns not needed, like x and y coordinates of hospital 
HospitalBedData = HospitalBedData.drop(['X','Y','HQ_ADDRESS1','HQ_STATE','COUNTY_NAME','STATE_NAME','STATE_FIPS','CNTY_FIPS'], axis=1)

#turn all NANs into 0
HospitalBedData = HospitalBedData.fillna(0)

#since im merging on FIPS, if a NAN value is present it would cause problems, so those are being removed
HospitalBedData = HospitalBedData[HospitalBedData['FIPS'].notna()]

#switch data types from floats to ints for ease
HospitalBedData = HospitalBedData.astype({'FIPS': 'int','NUM_LICENSED_BEDS':'int','NUM_STAFFED_BEDS':'int','PEDI_ICU_BEDS':'int'})

#view the final table
HospitalBedData.head()
```

### Below the COVID superset and the hospital data are being merged on FIPS 
#### Both datasets have FIPS but with a diffrent name, also repeated data was removed like duplicate FIPS and old index from hospital data

```python
#merge the superset and hosiptal data on FIPS, both have the value but have a diffrent name
HospitalBedSupersetMerge = pd.merge(SuperData,HospitalBedData,how="right",left_on="countyFIPS",right_on="FIPS")
#remove duplicate FIPS and the old index
HospitalBedSupersetMerge = HospitalBedSupersetMerge.drop(['OBJECTID','FIPS'], axis=1)
HospitalBedSupersetMerge.head()
```
