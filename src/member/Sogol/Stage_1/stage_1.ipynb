importing pandas and matplotlib to read the file and then plot the data 

```python
import pandas as pd 
import matplotlib.pyplot as plt

cases = pd.read_csv("../../../../data/stage_I/covid_confirmed_usafacts.csv", delimiter=",")
```

the state I have chosen to analyze is New York so I'm getting the confirmed cases data since last week 

```python
ny_cases = cases[cases['State'] == 'NY']
```

```python
x=list(ny_cases.columns[-7:])
y=list(ny_cases.sum())[-7:]
```

```python
plt.figure(figsize=(10,10))
plt.plot(x,y,linestyle='solid')
```

based on the plot above we see that in state of New York there has been always an increase, in other words, number of cases have been increasing everyday. If we go to detail, there has been no cases(stable) from day 5th of February 2021 to the 6th and then there were a lot of new cases since that day to February 7th and there were even more cases since February 7th to the 10th and number of cases has been increasing since day 10th of February to the 11th and definitelly it keeps going up everyday unless people get vaccinated.
