### Choose NC state to analyze

Using masking, I extracted all rows whose 'State' value is 'NC' to a new dataframe

```python
nc_data = super_dataset[super_dataset['State'] == 'NC']
nc_data
```

### Visualize the trend for the last 7 days

I extracted the seven columns for COVID-19 cases. Because we are dealing with time series data, it is generally a good practice to convert date like string to datetime. Here, I took the total sum of cases across all counties of NC, converted it into dictionary so that dates are keys and sum of cases for a particular key are the values.

Iterating through the dictionary items, I pull out the dates and number of cases as two different list. These lists are then passed as arguments to the plot_date function of matplotlib.pyplot. This gives us a beautiful visualize of the change in COVID-19 cases across the last 7 days

```python
# Get the column name for cases of the last 7 days
cases_cols = [i for i in list(nc_data.columns[-7*2:]) if 'cases' in i]

dates = []
case_vals = []
# Get the cases as series, find the sum, convert to dict and iterate
for k,v in nc_data[cases_cols].sum().to_dict().items():
    # remove '_cases' substring from column names, split by '-' and convert the list of int(s) to date
    dates.append(date(*[int(i) for i in k.replace("_cases", "").split("-")]))
    case_vals.append(v)
```

```python
# Give height and width argument to change the size of plot
plt.figure(figsize=(20, 10))
# pass dates and values and connect the dots with solid line
plt.plot_date(dates, case_vals, linestyle='solid')
# set labelsize to 18 for better display of text
plt.tick_params(labelsize=18)
# show the plot
plt.show()
```

### Load and display ACS Census Demographics Enrichment dataset

```python
dp5 = pd.read_csv("../../../../data/stage_I/Census_Demographics_ACS/ACSDP5Y2019.DP05_data_with_overlays_2021-02-11T145452.csv", delimiter=",", dtype=object)
dp5
```

### Preprocess data

I applied lambda function on 'GEO_ID' to take the last 5 characters and convert it into integer. I then added a new column named 'countyFIPS' to enrichment dataframe and rearranged the columns of the new dataframe

```python
dp5['countyFIPS'] = dp5[1:]['GEO_ID'].apply(lambda x: int(x[-5:]))
dp5['countyFIPS'].fillna(0, inplace=True)
dp5.drop(['GEO_ID'], axis = 1, inplace=True)
dp5 = dp5.astype({'countyFIPS': 'int32'})
dp5 = dp5[[dp5.columns[-1]] + list(dp5.columns[:-1])]
dp5
```

### Merge enrichment dataset with super_dataset

Because the two dataframe have 'countyFIPS' as the common column name, merging enrichment dataset with super_dataset will automatically use 'countyFIPS' as the intersection of keys from both dataframes.

```python
super_dataset.merge(dp5[1:])
```
