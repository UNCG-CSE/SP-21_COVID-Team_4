### Choose North Carolina state to analyze

Using masking, I extracted all rows whose 'State' value is 'NC' to a new dataframe

```python
nc_data = super_dataset[super_dataset['State'] == 'NC']
nc_data
```

### Visualize the trend

I extracted the COVID-19 confirmed cases and deaths seperately from superdataset. Here, I took the total sum of cases across all counties of NC, converted it into dictionary so that dates are keys and sum of cases for a particular key are the values.

Because we are dealing with time series data, it is generally a good practice to convert date like string to datetime format. Iterating through the index of total_cases Series, I converted date like strings to date. I also created two seperate list containing cases and deaths.

These lists are then passed as arguments to the plot_date function of matplotlib.pyplot. This gives us a beautiful visualization of the trend in COVID-19 cases and deaths across North Carolina.

```python
# Get all column names for cases
cases_cols = [i for i in list(nc_data.columns) if 'cases' in i]
# Use the list above to select cases only, and sum all datas across each columns
total_cases = nc_data[cases_cols].sum()

# Get all column names for deaths
deaths_cols = [i for i in list(nc_data.columns) if 'deaths' in i]
# Use the list above to select deaths only, and sum all datas across each columns
total_deaths = nc_data[deaths_cols].sum()
```

'total_cases' and 'total_deaths' give the **aggregated data** of confirmed cases and deaths **across entire North Carolina** from start of the case till the last date

```python
# Convert date like string to datetime.date format
dates = [date(*[int(i) for i in k.replace("_cases", "").split("-")]) for k in total_cases.index]

# Get the list of all cases and death count
cases_vals = list(total_cases.values)
deaths_vals = list(total_deaths.values)
```

```python
font_s = 20
# Give height and width argument to change the size of plot
plt.figure(figsize=(20, 10))
# pass dates and values and connect the dots with solid line
plt.plot_date(dates, cases_vals, linestyle='solid', markersize=1)
# set labelsize to 18 for better display of text
plt.tick_params(labelsize=font_s)
plt.title("\nTrend of confirmed cases in North Carolina\n", fontsize=font_s*1.5, fontweight="bold")
plt.xlabel('\nDate: 2020 - 2021', fontsize=font_s)
plt.ylabel('\nNumber of confirmed cases\n', fontsize=font_s)
# show the plot
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.plot_date(dates, deaths_vals, linestyle='solid', color='green', markersize=1)
plt.tick_params(labelsize=font_s)
plt.title("\nTrend of deaths in North Carolina\n", fontsize=font_s*1.5, fontweight="bold")
plt.xlabel('\nDate: 2020 - 2021', fontsize=font_s)
plt.ylabel('\nNumber of deaths\n', fontsize=font_s)
plt.show()
```

### Trend for the last 7 days

```python
plt.figure(figsize=(20, 10))
plt.plot_date(dates[-7:], cases_vals[-7:], linestyle='solid', markersize=10)
plt.tick_params(labelsize=font_s)
plt.title("\nTrend of confirmed cases in NC in the last week\n", fontsize=font_s*1.5, fontweight="bold")
plt.xlabel('\nDate', fontsize=font_s)
plt.ylabel('\nNumber of confirmed cases\n', fontsize=font_s)


plt.figure(figsize=(20, 10))
plt.plot_date(dates[-7:], deaths_vals[-7:], linestyle='solid', color='green', markersize=10)
plt.tick_params(labelsize=font_s)
plt.title("\nTrend of deaths in NC in the last week\n", fontsize=font_s*1.5, fontweight="bold")
plt.xlabel('\nDate', fontsize=font_s)
plt.ylabel('\nNumber of deaths\n', fontsize=font_s)
plt.show()
```

### Load and display ACS Census Demographics Enrichment dataset

```python
dp5 = pd.read_csv("../../../../data/stage_I/Census_Demographics_ACS/ACSDP5Y2019.DP05_data_with_overlays_2021-02-11T145452.csv", delimiter=",", dtype=object)
dp5
```

### Preprocess data

I applied lambda function on 'GEO_ID' to take the last 5 characters and convert it into integer. I then added a new column named 'countyFIPS' to enrichment dataframe and rearranged the columns of the new dataframe

```python
dp5['countyFIPS'] = dp5[1:]['GEO_ID'].apply(lambda x: int(x[-5:]))
dp5['countyFIPS'].fillna(0, inplace=True)
dp5.drop(['GEO_ID'], axis = 1, inplace=True)
dp5 = dp5.astype({'countyFIPS': 'int32'})
dp5 = dp5[[dp5.columns[-1]] + list(dp5.columns[:-1])]
dp5
```

### Merge enrichment dataset with super_dataset

Because the two dataframe have 'countyFIPS' as the common column name, merging enrichment dataset with super_dataset will automatically use 'countyFIPS' as the intersection of keys from both dataframes.

```python
new_dataset = super_dataset.merge(dp5[1:])
new_dataset
```
